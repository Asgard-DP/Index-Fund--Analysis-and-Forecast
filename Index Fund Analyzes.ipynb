# Analyze the data


```python
# Initial imports

import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path

%matplotlib inline
```

```python
# get the data

XLRE_df = pd.read_csv(
    Path('../Project 1/Resouces/XLRE.csv'), index_col="Date", infer_datetime_format=True, parse_dates=True)

XLK_df = pd.read_csv(
    Path('../Project 1/Resouces/XLK.csv'), index_col="Date", infer_datetime_format=True, parse_dates=True)

XLF_df = pd.read_csv(
    Path('../Project 1/Resouces/XLF.csv'), index_col="Date", infer_datetime_format=True, parse_dates=True)

BTC_df = pd.read_csv(
    Path('../Project 1/Resouces/BTC-USD.csv'), index_col="Date", infer_datetime_format=True, parse_dates=True)


# display top 5 and bottom 5 days in last 5 years 
display(XLRE_df.head())
display(XLRE_df.tail())


display(XLK_df.head())
display(XLK_df.tail())

display(XLF_df.head())
display(XLF_df.tail())

display(BTC_df.head())
display(BTC_df.tail())
```

```python
#Drop All Columns BUT Close 

XLRE_close_df = XLRE_df.drop(['Open','High','Low','Adj Close','Volume'], axis=1)

XLK_close_df = XLK_df.drop(['Open','High','Low','Adj Close','Volume'], axis=1)

XLF_close_df = XLF_df.drop(['Open','High','Low','Adj Close','Volume'], axis=1)

BTC_close_df = BTC_df.drop(['Open','High','Low','Adj Close','Volume'], axis=1)

display(XLRE_close_df.head())
display(XLK_close_df.head())
display(XLF_close_df.head())
display(BTC_close_df.head())
```

```python
# calculate daily returns 

XLRE_daily_returns = XLRE_close_df.pct_change()

XLK_daily_returns = XLK_close_df.pct_change()

XLF_daily_returns = XLF_close_df.pct_change()

BTC_daily_returns = BTC_close_df.pct_change()

# Rename the close to match the Fund. 

XLRE_daily_returns.columns = ['XLRE']

XLK_daily_returns.columns = ['XLK']

XLF_daily_returns.columns = ['XLF']

BTC_daily_returns.columns = ['BTC']

# display DF

display(XLRE_daily_returns.head())
display(XLK_daily_returns.head())
display(XLF_daily_returns.head())
display(BTC_daily_returns.head())

```

```python
# DROP NULL 
XLRE_daily_returns.dropna(inplace=True)
XLK_daily_returns.dropna(inplace=True)
XLF_daily_returns.dropna(inplace=True)
BTC_daily_returns.dropna(inplace=True)
```

```python
combined_daily_returns_df = pd.concat([XLRE_daily_returns,XLK_daily_returns,XLF_daily_returns,BTC_daily_returns], axis='columns', join='inner')

display(combined_daily_returns_df.head())
```

# Conduct Quantitative Analysis
### Calculate and Plot cumulative returns.

```python
# Calculate cumulative returns of all portfolios
cumulative_returns = (1 + combined_daily_returns_df).cumprod() - 1

# Plot cumulative returns
cumulative_returns.plot(figsize=(10,5),
                       title= " ")

cumulative_returns.describe()
```

```python
# record of the dataframe multiplied by 100 is giving us the percentage change of the stock Index Fund for our entire period 12/21/2022-12/21/2017

cum_return_entire_period = cumulative_returns.tail(1)

print('XLK had the best commulative return in the previous 5 years with a 97.85% return in index fund. the largest retuns percentnage incresase from 2018 was 178% in 2022')

cum_return_entire_period * 100
```

# Sharpe Ratios

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
sharpe_ratio = combined_daily_returns_df.mean()/combined_daily_returns_df.std()
asr = sharpe_ratio*252**.5
asr
```

```python
# Visualize the sharpe ratios as a bar plot

asr.plot(kind='bar')
```

```python
print('XLK has the best Sharpy ratio at 0.63 compaired to BTC, XLF, and XLRE funds . XLK would have been the best investment to make in the preiovious 5 years, followed by BTC , XLF, XLRE.')
```
